#Sigmoid Function

```python
#import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function define
def sigmoid(t):
  return 1/(1+np.exp(-t))
```

```python
#Test
print(sigmoid(0))
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
```

```python
plt.figure(figsize=(12 , 5), dpi=300)
plt.plot(x, y, lw=2, linestyle='-', color='b')
plt.grid(color='r')
plt.title('sigmoid function')
plt.xlabel('independent variable')
plt.ylabel('sigmoid values')
plt.xticks(rotation=45)
plt.show()
```
